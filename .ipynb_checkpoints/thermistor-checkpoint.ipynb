```python
# R2 = R1/exp(Betha*(1/T1-1/T2)) #in kelvins
T = linspace(20, 100, 200)
R = 1500/exp(3900*(1/298.15-1/(T+273.15)))
```

```python
# R-T graph for K164NK001.5 termistor
plot(T,R)
grid(True)
```

```python
#8bit ADC (Arduino)value vs Temperature
I=0.002
ADC=(5-I*R)/5*1024
plot(T, ADC)
grid(True)
```

```python
#Alternative use as pasive protection in two transistor current source circuit
Ren=820+0
Vin=12
Vout=R/(Ren+R)*Vin
plot(T, Vout)
Vgate=3.8
plot([T[0], T[-1]], [Vgate, Vgate])
```
